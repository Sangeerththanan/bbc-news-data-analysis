```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Step 1: Define the URL
url = "https://www.bbc.com/news"

# Step 2: Fetch the webpage content
response = requests.get(url)
if response.status_code == 200:
    print("Successfully fetched the webpage!")
else:
    print(f"Failed to fetch webpage. Status code: {response.status_code}")

# Step 3: Parse the webpage content
soup = BeautifulSoup(response.content, "html.parser")

# Step 4: Extract data
news_data = []

# Loop through each news card
for card in soup.find_all("div", {"data-testid": "anchor-inner-wrapper"}):
    # Extract article link
    link_tag = card.find("a", {"data-testid": "internal-link"})
    link = link_tag['href'] if link_tag else "N/A"
    full_link = f"https://www.bbc.com{link}"  # Making the URL absolute
    
    # Extract headline
    headline_tag = card.find("h2", {"data-testid": "card-headline"})
    headline = headline_tag.get_text(strip=True) if headline_tag else "N/A"
    
    # Extract description
    description_tag = card.find("p", {"data-testid": "card-description"})
    description = description_tag.get_text(strip=True) if description_tag else "N/A"
    
    # Extract last updated
    updated_tag = card.find("span", {"data-testid": "card-metadata-lastupdated"})
    last_updated = updated_tag.get_text(strip=True) if updated_tag else "N/A"
    
    # Extract category/tag
    tag_tag = card.find("span", {"data-testid": "card-metadata-tag"})
    tag = tag_tag.get_text(strip=True) if tag_tag else "N/A"
    
    # Extract image URL
    img_tag = card.find("img")
    img_url = img_tag['src'] if img_tag else "N/A"

    # Append to the list
    news_data.append({
        "Headline": headline,
        "Description": description,
        "Last Updated": last_updated,
        "Category": tag,
        "Image URL": img_url,
        "Article Link": full_link
    })

# Step 5: Save the data to a Pandas DataFrame
df = pd.DataFrame(news_data)

# Step 6: Export to a CSV file
df.to_csv("bbc_news_data.csv", index=False)
print("Data saved to bbc_news_data.csv!")
```

```python
import pandas as pd
import os

# Define the file path
file_path = "bbc_news_data.csv"

# Step 1: Check if the file exists
if not os.path.exists(file_path):
    print(f"Error: The file '{file_path}' does not exist.")
else:
    # Step 2: Load the CSV file
    data = pd.read_csv(file_path)
    
    # Step 3: Check if the necessary columns exist
    required_columns = ["Headline", "Article Link", "Image URL"]
    if not all(col in data.columns for col in required_columns):
        print("Error: One or more required columns are missing in the CSV file.")
    else:
        # Step 4: Remove duplicate rows
        data.drop_duplicates(subset=["Headline", "Article Link"], inplace=True)

        # Step 5: Remove rows with missing or placeholder data
        data = data[data["Headline"].notna() & (data["Article Link"] != "N/A")]

        # Step 6: Replace placeholder image URLs with a default image if necessary
        data["Image URL"] = data["Image URL"].apply(
            lambda x: "https://example.com/no-image.png" if x == "/bbcx/grey-placeholder.png" else x
        )

        # Step 7: Normalize Article Links (add "https://www.bbc.com" if not present)
        data["Article Link"] = data["Article Link"].apply(
            lambda x: x if x.startswith("https://") else f"https://www.bbc.com{x}"
        )

        # Step 8: Save the cleaned data to a new CSV file
        cleaned_file_path = "bbc_news_data_cleaned.csv"
        data.to_csv(cleaned_file_path, index=False)

        # Step 9: Print success message
        print(f"Cleaned data saved to {cleaned_file_path}")
```

```python
# 1. Category Distribution Analysis
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv("bbc_news_data_cleaned.csv")

# Handle missing or empty categories by replacing them with 'Unknown'
df['Category'] = df['Category'].fillna('Unknown')

# Check for category counts
category_counts = df["Category"].value_counts()

# Plot the category distribution
plt.figure(figsize=(10,6))
ax = category_counts.plot(kind="bar", color="skyblue")
plt.title("Category Distribution of BBC News Articles")
plt.xlabel("Category")
plt.ylabel("Number of Articles")
plt.xticks(rotation=45, ha='right')

# Add count labels on top of the bars
for i, value in enumerate(category_counts):
    ax.text(i, value + 0.2, str(value), ha='center', va='bottom', fontsize=10)

plt.tight_layout()

# Show the plot
plt.show()
```
